```python
from datetime import date, timedelta
from garminconnect import Garmin
from pathlib import Path
import pandas as pd
import os


# Paths
TOKEN_DIR = Path("/content/drive/MyDrive/Moje projekty/Garmin Connect/Garmin tokens")
EXPORT_DIR = Path("/content/drive/MyDrive/Moje projekty/Garmin Connect/Garmin exports")
os.makedirs(EXPORT_DIR, exist_ok=True)


# Login
garmin = Garmin()
garmin.login(str(TOKEN_DIR))


# Login Verification
try:
    garmin.get_user_profile()
except Exception as e:
    raise Exception(f"Login failed! Check your tokens. Details: {e}")


# Start & end date
start = date(2020, 12, 21)
end = date.today() - timedelta(days=1)


# Fetch data
rows = []
current = start

while current <= end:
    iso = current.isoformat()
    try:
        summary = garmin.get_user_summary(iso)
        steps = summary.get("totalSteps") or summary.get("steps") or 0
    except Exception:
        steps = 0

    rows.append({"Date": iso, "Steps": steps})
    current += timedelta(days=1)

df = pd.DataFrame(rows)

# Save CSV
file_name = f"steps_export_{start.isoformat()}_to_{end.isoformat()}.csv"
file_path = EXPORT_DIR / file_name

df.to_csv(file_path, index=False)

print("\n=== Export summary ===")
print(f"Selected option: {option_description}")
print(f"Date range: {start.isoformat()} → {end.isoformat()}")
print(f"File saved to: {file_path}")
```
