```python
"""
Garmin Connect Token Retriever

This script logs into Garmin Connect using user-provided credentials, handles Multi-Factor Authentication (MFA) if required, and securely
saves authentication tokens locally for future API requests.

"""


from garminconnect import Garmin
from getpass import getpass
from pathlib import Path


# Directory to store authentication tokens
TOKEN_DIR = Path("garmin_tokens")
TOKEN_DIR.mkdir(parents=True, exist_ok=True)


# Prompt user for Garmin login credentials securely
email = input("Enter your Garmin email: ")
password = getpass("Enter your Garmin password (input hidden): ")


garmin = Garmin(email=email, password=password, return_on_mfa=True)

try:
    login_result = garmin.login()
except Exception as e:
    print(f"Login failed: {e}")
    exit(1)


# Handle Multi-Factor Authentication (MFA) if required
if isinstance(login_result, tuple) and login_result[0] == "needs_mfa":
    mfa_code = input("Enter MFA code from Garmin Authenticator or SMS: ")
    garmin.resume_login(login_result[1], mfa_code)


try:
    garmin.garth.dump(TOKEN_DIR)
    print(f"Authentication tokens successfully saved in: {TOKEN_DIR}")
except Exception as e:
    print(f"Failed to save tokens: {e}")
```
